```python
import os 

fileList = os.listdir('data/daylightingBand/images/train')
for file in fileList:
    file_name = os.path.splitext(file)[0]
    file_path = os.path.join('data/daylightingBand/labels/train', file_name + '.txt')
    try: 
        with open(file_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                line = line.strip().split()
                print(line)
    except FileNotFoundError:
        print(file_path, 'not found')
```
